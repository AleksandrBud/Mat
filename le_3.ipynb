```python
s_1 = input('Введите вектор: ')
v_1 = np.fromstring(s_1, dtype=float, sep=',' )
print(v_1)
sum_sqr = 0
for i in v_1:
    sum_sqr = sum_sqr + i**2
len_v_1 = np.sqrt(sum_sqr)
print(len_v_1)
```

    2)  Пусть задана плоскость: A1x + B1y + C1z + D1 = 0
        и прямая: 
        (x-x1)/(x2-x1)=(y-y1)/(y2-y1)=(z-z1)/(z2-z1)
        Как узнать, принадлежит прямая плоскости или нет?
        
    Ответ: Если A1x1 + B1y1 + C1z1 + D1 = 0 и A1x2 + B1y2 + C1z2 + D1 = 0, то прямая принадлежит плоскости.

```python
#5. Задание (в программе)
#    1) Нарисуйте трехмерный график двух параллельных плоскостей.
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-100, 5, 0.5)
Y = np.arange(-100, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z = X + Y
ax.plot_wireframe(X, Y, Z)
X2 = np.arange(-100, 5, 0.5)
Y2 = np.arange(-100, 5, 0.5)
X2, Y2 = np.meshgrid(X2, Y2)
Z2 = X + Y + 100
ax.plot_wireframe(X2, Y2, Z2)
show()
```

```python
# 2) Нарисуйте трехмерный график двух любых поверхностей второго порядка.
fig = figure()
ax = Axes3D(fig)
X = np.arange(-100, 100, 0.1)
Y = np.arange(-100, 100, 0.1)
X, Y = np.meshgrid(X, Y)
Z = X**2 - Y**2
ax.plot_wireframe(X, Y, Z)
show()
```
